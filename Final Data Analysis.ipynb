```python
Image("Output/Tourism Employment in Canada.png")
```

# Observations
* Tourism provides significant employment to young people. Tourism sector has been one of the significant sources of employment  for newcomers to Canada. 
* The Canadian tourism industry accounted for approximately 752.2 thousand jobs to the country's economy in Q4 2019.
* Canada's unemployment rate in the tourism sector was approximately 5.6% in February of 2020, which is 0.8% higher than the previous year.
* On a provincial basis, tourism unemployment rates ranged from 4.3% in Manitoba to 15.5% in Prince Edward Island.
* One of the main reasons for the rise in unemployment in the tourism sector was due to the decrease in tourism spending in Canada in the wake of continued travel restrictions and physical distancing measures throughout Canada
* In the second quarter of 2020, tourism spending by international visitors came to a halt as the border remained closed to non-essential foreign travellers and hence we see a sharp decline in tourism employment
